```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from matplotlib.dates import DateFormatter
from matplotlib.ticker import MultipleLocator
```

```python
data = pd.read_excel('PM25.xlsx',sheet_name =1)
```

```python
## calculate the linear regression

len(ns)
t = np.arange(1,390,1)
```

```python
slope, intercept, r_value, p_value, std_err = linregress(t,NS)

print("Slope:", slope)
print("Intercept:", intercept)
print("Correlation coefficient (r):", r_value)
print("P-value:", p_value)
print("Standard error:", std_err)
```

```python
y=[]
for i in t:
    y.append(slope*i+intercept)

len(y)
```

```python
from matplotlib.dates import DateFormatter
from matplotlib.ticker import MultipleLocator

plt.plot(ns,NS)
plt.plot(ns,y,'--',label=('-0.05*x+50.85'))

# Label the x-axis
# plt.xlabel('Date')
# Set the x-axis tick locator to show labels every 5 data points
plt.gca().xaxis.set_major_locator(MultipleLocator(50))

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

plt.title('PM2.5 concentration at NS station')
plt.ylabel('conc in ug/m3')
plt.legend()
```
