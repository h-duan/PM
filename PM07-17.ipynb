```python
##  NH4+ annual concentration GZ

NH4ave = []  # annual average of NH value

for i in range(len(GZ)):
    #print(i)
    tgc = GZ[i].loc[1:,'NH4+']
    NH4ave.append(tgc.mean())

# calculate the regression line
array = np.arange(len(NH4ave))
t = np.arange(2008,2018)

slope, intercept, r_value, p_value_nh4gz, std_err = linregress(array, NH4ave)

ygz=[]  # calculated value for gz using linear regression
for i in range(len(t)):
    ygz.append(slope*i+intercept)

print(p_value_nh4gz)
```

```python
# adjust the list to the same length with annual mean NH4ave
#ygz[-1]=np.nan
ygz[-2]=np.nan
```

```python
print('r value  '+str(r_value))
y = NH4ave
ypred = slope*array+intercept
ss_tot = np.sum((y[:-3] - np.mean(y[:-3]))**2)
ss_res = np.sum((y[:-1] - ypred[:-1])**2)
r_squared = 1 - (ss_res / ss_tot)

# Print the R-squared value
print("R-squared:", r_squared)
```
